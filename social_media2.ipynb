```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://soax.com/research/time-spent-on-social-media"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.find('tbody')

    years = []
    minutes_per_day = []
    changes = []

    for row in table.find_all('tr'):
        cells = row.find_all('td')
        if cells:  
            years.append(cells[0].text.strip())
            minutes_per_day.append(cells[1].text.strip())
            changes.append(cells[2].text.strip())

    data = {
        'Year': years,
        'Minutes / day': minutes_per_day,
        'Change': changes
    }
    mpd = pd.DataFrame(data)
```
