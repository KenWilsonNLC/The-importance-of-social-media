```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://soax.com/research/time-spent-on-social-media"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    tables = soup.find_all('table')
    
    if len(tables) >= 2:
        second_table = tables[1]
        rows = second_table.find_all('tr')
        
        data = []
        
        for row in rows[1:]:
            cols = row.find_all('td')
            platform = cols[0].text.strip()
            duration = cols[1].text.strip()
            data.append((platform, duration))
        
        platform = pd.DataFrame(data, columns=['Platform', 'Average Session Duration (Minutes)'])
        
        platform['Average Session Duration (Minutes)'] = pd.to_timedelta(platform['Average Session Duration (Minutes)']).dt.total_seconds() / 60
```
